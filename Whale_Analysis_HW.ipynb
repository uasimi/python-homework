# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:

* `whale_returns.csv`: Contains returns of some famous "whale" investors' portfolios.

* `sp500_history.csv`: Contains historical closing prices of the S&P 500 Index.

## Whale Returns

Read the Whale Portfolio daily returns and clean the data

```python
# Read whale returns
whale_path = Path("Resources/whale_returns.csv")
whale_returns = pd.read_csv(whale_path, index_col='Date', infer_datetime_format=True, parse_dates=True)
whale_returns
```

```python
# count nulls
whale_returns.isnull().sum()
```

```python
# drop nulls 
whale_returns.dropna(inplace = True)
whale_returns
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
# read algo returns 
algo_path = Path("Resources/algo_returns.csv")
algo_returns = pd.read_csv(algo_path, index_col='Date', infer_datetime_format=True, parse_dates=True)
algo_returns
```

```python
# count nulls
algo_returns.isnull().sum()
```

```python
# drop nulls
algo_returns.dropna(inplace = True)
algo_returns
```

## S&P 500 Returns

Read the S&P 500 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# read sp500 prices
sp500_path = Path("Resources/sp500_history.csv")
sp500_returns = pd.read_csv(sp500_path, index_col='Date', infer_datetime_format=True, parse_dates=True)
sp500_returns
```

```python
# data info
sp500_returns.info()
```

```python
# fix dtypes
sp500_returns["Close"] = sp500_returns["Close"].str.replace("$", "")
sp500_returns["Close"] = sp500_returns["Close"].astype("float")
sp500_returns["Close"]
```

```python
# calc daily returns
sp500_daily_returns = sp500_returns.pct_change()
sp500_daily_returns
```

```python
# drop nulls
sp500_daily_returns = sp500_daily_returns.dropna()
```

```python
# rename 'close' 
sp500_daily_returns.columns = ["S&P 500 Returns"]
sp500_daily_returns
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# concat returns to single df as columns
combined_returns = pd.concat([whale_returns, algo_returns, sp500_daily_returns], axis="columns", join="inner")
combined_returns.head()
```

# Conduct Quantitative Analysis

In this section, you will calculate and visualize performance and risk metrics for the portfolios.

## Performance Anlysis

#### Calculate and Plot the daily returns.

```python
# plot all daily returns
combined_returns.plot(kind="line", title = "Returns", figsize = (30,15), subplots = True, ylim = (-.10, .10))
```

## Risk Analysis

Determine the _risk_ of each portfolio:

1. Create a box plot for each portfolio. 
2. Calculate the standard deviation for all portfolios
4. Determine which portfolios are riskier than the S&P 500
5. Calculate the Annualized Standard Deviation

### Create a box plot for each portfolio

```python
# box plot to determine risk
combined_returns.plot(kind = "box", figsize = (20,10))
```

### Calculate Standard Deviations

```python
# calc daily std of all portfolios
combined_std = combined_returns.std()
combined_std
```

### Determine which portfolios are riskier than the S&P 500

```python
# Calculate  the daily std of S&P 500
sp500_std = combined_returns["S&P500 Returns"].std()

# Determine which portfolios are riskier than the S&P 500
riskier_portfolio = combined_std.index[combined_std > sp500_std].tolist()
print(f"({riskier_portfolio} are risker then S&P 500 due to a higher SD")
```

### Calculate the Annualized Standard Deviation

```python
# Calculate the annualized std (252 trading days)
annualized_std = combined_std * np.sqrt(252)
annualized_std
```

### Calculate and plot rolling `std` for all portfolios with 21-day window

```python
# Calculate the rolling std for all portfolios using a 21-day window
std_return_21day = combined_returns.rolling(window=21).std().dropna()
# Plot the rolling std
std_return_21day.plot(title = "Rolling Standard Deviation: 21-day", figsize = (20,10))
```

### Calculate and plot the correlation

```python
# Calc correlation
correlation = combined_returns.corr()
# Display correlation matrix
sns.heatmap(correlation, vmin = -1, vmax = 1)
```

### Calculate and Plot Beta for a chosen portfolio and the S&P 500

```python
# Calc covariance of a single portfolio
algo1_cov = combined_returns["Algo 1"].cov(combined_returns["S&P500 Returns"])
# Calc variance of S&P 500
sp500_var = combined_returns["S&P500 Returns"].var()
# Compute beta
algo1_beta = algo1_cov / sp500_var
# Plot beta trend
algo1_cov21 = combined_returns["Algo 1"].rolling(window=21).cov(combined_returns["S&P500 Returns"])
sp500_var21 = combined_returns["S&P500 Returns"].rolling(window=21).var()
algo1_beta21 = algo1_cov21 / sp500_var21
algo1_beta21.plot()
```

```python
# Read 1st stock data
goog_path = Path("Resources/goog_historical.csv")
goog_df = pd.read_csv(goog_path, index_col='Trade DATE', infer_datetime_format=True, parse_dates=True)
goog_df.rename(columns = {"NOCP": "GOOG"}, inplace = True)
goog_df.drop(columns=["Symbol"], inplace = True)
goog_df
```

```python
# Read 2nd stock data
aapl_path = Path("Resources/aapl_historical.csv")
aapl_df = pd.read_csv(aapl_path, index_col='Trade DATE', infer_datetime_format=True, parse_dates=True)
aapl_df.rename(columns = {"NOCP": "AAPL"}, inplace = True)
aapl_df.drop(columns=["Symbol"], inplace = True)
aapl_df
```

```python
# Read 3rd stock data 
cost_path = Path("Resources/cost_historical.csv")
cost_df = pd.read_csv(cost_path, index_col='Trade DATE', infer_datetime_format=True, parse_dates=True)
cost_df.rename(columns = {"NOCP": "COST"}, inplace = True)
cost_df.drop(columns=["Symbol"], inplace = True)
cost_df
```

```python
# Combine stocks in a DataFrame
combined_df = pd.concat([goog_df, aapl_df, cost_df], axis="columns", join="inner")
```

```python
# Reset Date index
combined_df.sort_index(inplace = True)
combined_df.head()
```

```python
# Reorganize portfolio data by having a column per symbol
# Organized when csv was read in
#combined_df
```

```python
# Calc daily returns
daily_returns_df = combined_df.pct_change()

# Drop NAs
daily_returns_df.dropna(inplace = True)

# Display sample data
daily_returns_df.head()
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Set weights
weights = [1/3, 1/3, 1/3]

# Calc portfolio return
daily_returns_df["My Portfolio"] = daily_returns_df.dot(weights)
# Display sample data
daily_returns_df.head()
```

## Join your portfolio returns to the DataFrame that contains all of the portfolio returns

```python
# Join your returns DataFrame to the original returns DataFrame
total_returns = pd.concat([combined_returns, daily_returns_df["My Portfolio"]], axis="columns", join="inner")
total_returns.rename(columns = {"0": "My Portfolio"}, inplace = True)
total_returns.head()
```

```python
# compare dates where return data exists for all the stocks (drop NaNs)
total_returns.dropna(inplace = True)
```
